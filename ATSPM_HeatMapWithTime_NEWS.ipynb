```python
# Marker Lat and Lon
df_oneday['Latitude_M'] = df_oneday['Latitude']
df_oneday['Longitude_M'] = df_oneday['Longitude']
```

```python
for index, row in df_oneday.iterrows():
    lat, lon, direction_code = row['Latitude'], row['Longitude'], str(row['SignalID_Direc'])[-1]
    #detector = row['DetID']
    #parity = int(detector.split('_', 2)[1])
    
    if direction_code == "N":
        marker_lat = lat - 0.0010 
        marker_lon = lon #- 0.0002 
    elif direction_code == "S":
        marker_lat = lat + 0.0010
        marker_lon = lon #- 0.0002 
    elif direction_code == "E":
        marker_lat = lat #- 0.0002 
        marker_lon = lon - 0.0010
    elif direction_code == "W":
        marker_lat = lat #- 0.0002 
        marker_lon = lon + 0.0010
    else:
        print("Error, invalid direction code")
        
    df_oneday.set_value(index,'Latitude_M',marker_lat)
    df_oneday.set_value(index,'Longitude_M',marker_lon)
```

```python
df_oneday_start = df_oneday[df_oneday['RelTime']==df_oneday['RelTime'].min()]
df_oneday_start.head()
```

```python
#max_val = 350 #df_oneday['ON_count'].max()
max_val = df_oneday['ON_count_mean'].quantile(0.995)
min_val = df_oneday['ON_count_mean'].min()

df_oneday['Weight'] = (df_oneday['ON_count_mean'] - min_val)/(max_val - min_val)
df_oneday['Weight'] = np.clip(df_oneday['Weight'], 0, 1)
df_oneday
```

```python
import folium # contains leaflet wrappers for python
from folium import plugins
from IPython.display import IFrame
from folium.plugins import MarkerCluster



map_orlando = folium.Map(location=[28.697794, -81.326950],
                            zoom_start=12) 
kw = {
    'prefix': 'fa',
    'color': 'gray',
    'icon': 'arrow-up'
}

mc = MarkerCluster()

angles = {
          "N" :   0,
          "S" : 180, 
          "E" :  90, 
          "W" : 270
         }

for index, row in df_oneday_start.iterrows():
    
    marker_lat = row['Latitude_M']
    marker_lon = row['Longitude_M']
    direction_code = str(row['SignalID_Direc'])[-1]
    detid = row['SignalID_Direc']
    
    # puting on map
    angle = angles[direction_code]
    icon = folium.Icon(angle=angle,
                       **kw)
    mc.add_child(folium.Marker(location=[marker_lat, marker_lon], 
                  icon=icon, 
                  tooltip=str(detid)))
                 
    
    
   
    
map_orlando.add_child(mc)
map_orlando.save(os.path.join('Rotate_icon.html'))

map_orlando
```
